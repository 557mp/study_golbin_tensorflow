```python
import numpy as np

x1 = np.random.uniform(0,10,100)
y  = x1*0.5 + np.random.normal(0,1,100) -1

# print(x1.shape)
# print(x1)
# print(y.shape)
# print(y)
# print(y.mean())

y = y.reshape(y.size,1)

# print(y.shape)
# print(y)
```

```python
X = np.c_[np.ones(x1.shape[0]), x1]
# print(X.shape)
# print(X)

import matplotlib.pyplot as plt

# print(X[:,1])
# print(y)
plt.scatter(X[:,1], y, s=10, c='r', marker='o') #입력값x, 입력값 y, 사이즈, 색, 마커모양
plt.show()
```

```python
# print(X.shape)
# print(y.shape)
# print(X)
# print(y)
theta = np.array([[0],[0]])
# print(theta)
print(X.dot(theta))

h = X.dot(theta)
J= np.sum(np.square(h-y))/(2*y.size)

print(J)
```
